```python
from langchain_core.messages import HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-pro-vision")

message = HumanMessage(
    content=[
        {
            "type": "text",
            "text": """Given the XKCD comic analyze it in detail. 
                    Consider the following aspects:
                      Visuals: Describe the comic's visual elements, including the characters, setting, and any objects or symbols. How do these visuals contribute to the overall humor or message of the comic?
                      Text: Analyze the text in the comic, including dialogue, captions, and any labels. How does the text interact with the visuals? What kind of language is used, and how does it contribute to the comic's tone and meaning?
                      Humor: Explain the source of humor in the comic. Is it based on wordplay, visual gags, irony, or a combination of these elements? How does the humor relate to the comic's message?
                      Message: Identify the underlying message or theme of the comic. Is it making a statement about science, technology, society, or human behavior? How does the comic use humor to convey its message?
                      Context: Consider the comic's publication date and any relevant events or trends at the time. Does the comic reference any specific cultural or scientific knowledge? How does this context affect the comic's meaning and humor?
                    Additionally, you can also discuss:
                      Your personal interpretation of the comic and how it resonates with you.
                      Whether the comic's humor is effective and whether its message is clear and impactful.
                      How the comic exemplifies the unique style and humor of XKCD"""
        },
        {"type": "image_url", "image_url": "https://www.explainxkcd.com/wiki/images/c/cc/ai_box_experiment.png"},
    ]
)

response = llm.invoke([message])

to_markdown(response.content)
```
